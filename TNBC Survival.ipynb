```python
import os
import pandas as pd
```

```python
OUTDIR = 'processed-data/'
```

```python
gdc_parsed_df = pd.read_csv(os.path.join(OUTDIR, '20180823_release_12.0.tsv'),
                            sep='\t', dtype=str)
gdc_parsed_df['os_censored'] = gdc_parsed_df['os_censored'].astype(str)
gdc_parsed_df['os_time'] = gdc_parsed_df['os_time'].apply(lambda x: pd.to_numeric(x, errors='coerce'))
gdc_parsed_df['os_time_years'] = gdc_parsed_df['os_time'] / 365.
gdc_parsed_df['os_time_years'] = gdc_parsed_df['os_time_years'].round(decimals=2)
gdc_parsed_df
```

### Quick comparison to Mustafas TCGA table

```python
brca_infile = 'Breast_TCGA_BRCA_Basal.csv'
brca_df = pd.read_csv(brca_infile)
brca_df.drop_duplicates(inplace=True)
brca_df
```

```python
merged_df = brca_df.merge(gdc_parsed_df, left_on='Patient', right_on='barcode')
```

```python
ax1 = merged_df.plot.scatter(x='Days', y='os_time', c='DarkBlue')
ax1.set_xlabel("Mustafa TCGA")
ax1.set_ylabel("Rahul TCGA")
```
