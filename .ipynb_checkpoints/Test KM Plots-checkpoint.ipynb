```python
import os
import pandas as pd
```

```python
OUTDIR = 'processed-data/'
```

```python
gdc_parsed_df = pd.read_csv(os.path.join(OUTDIR, '20180823_release_12.0.tsv'),
                            sep='\t', dtype=str)
gdc_parsed_df['os_censored'] = gdc_parsed_df['os_censored'].astype(str)
gdc_parsed_df['os_time'] = gdc_parsed_df['os_time'].apply(lambda x: pd.to_numeric(x, errors='coerce'))
gdc_parsed_df['os_time_years'] = gdc_parsed_df['os_time'] / 365.
gdc_parsed_df['os_time_years'] = gdc_parsed_df['os_time_years'].round(decimals=2)
gdc_parsed_df
```

# Lets test in GBM

```python
comparison_df = pd.read_csv('survival-plot2.tsv', sep='\t')
comparison_df
```

```python
test_df = gdc_parsed_df[(gdc_parsed_df['uuid'].isin(comparison_df['id'].unique())) &
                        (gdc_parsed_df['os_time'] > 0)]

test_male_df = test_df[test_df['gender'] == 'MALE']
test_male_df.sort_values(by=['os_time_years'], inplace=True)

test_female_df = test_df[test_df['gender'] == 'FEMALE']
test_female_df.sort_values(by=['os_time_years'], inplace=True)
```

```python
test_female_df
```

```python
timepoints = [0] + sorted(test_female_df['os_time_years'].unique())
cum_prob = 1

cum_probs = [1]

r = test_female_df.shape[0]

index = 0
end_index = len(timepoints) - 1
while index < end_index:
    start = timepoints[index]
    end = timepoints[index + 1]
    
    c_dict = test_female_df[(test_female_df['os_time_years'] >= start) &
                            (test_female_df['os_time_years'] <= end)]['os_censored'].value_counts()
    c = 0 if 'True' not in c_dict else c_dict['True']
    
    n = r-c
    
    d_dict =test_female_df[(test_female_df['os_time_years'] >= start) &
                           (test_female_df['os_time_years'] <= end)]['vital_status'].value_counts()
    d = 0 if 'Dead' not in d_dict else d_dict['Dead']
    
    print r, c, n, d, start, end
    
    prob = ((n - d) / float(n)) * cum_prob
    
    cum_prob = prob
    
    cum_probs.append(prob)
    print prob
    
    index += 1
    r = n-d
```
